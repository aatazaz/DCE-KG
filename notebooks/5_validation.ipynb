```python
import pandas as pd

triples = pd.read_csv("../data/processed/triples.csv")
extracted_set = set(zip(triples["head"], triples["relation"], triples["tail"]))

# Mock gold standard (replace with real DrugCentral data)
gold = {("Donepezil", "treats", "Alzheimer's disease")}

tp = len(extracted_set.intersection(gold))
fp = len(extracted_set - gold)
fn = len(gold - extracted_set)

precision = tp / (tp+fp) if (tp+fp)>0 else 0
recall = tp / (tp+fn) if (tp+fn)>0 else 0
f1 = 2 * precision * recall / (precision + recall) if (precision+recall)>0 else 0

print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}")
```
