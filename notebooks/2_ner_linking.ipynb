```python
import sys
sys.path.append("..")
from kg.entity_extractor import EntityExtractor
import pandas as pd
import os

extractor = EntityExtractor(bioportal_api_key=None)
all_entities = []

for fname in os.listdir("../data/raw"):
    if fname.endswith(".txt"):
        # Force UTF-8 encoding and ignore errors
        with open(f"../data/raw/{fname}", "r", encoding="utf-8", errors="ignore") as f:
            text = f.read()
        print(f"Processing {fname}... (text length: {len(text)})")
        entities = extractor.extract_entities(text)
        for e in entities:
            e["source_file"] = fname
            all_entities.append(e)

# Save to CSV
df = pd.DataFrame(all_entities)
if not df.empty:
    df.to_csv("../data/processed/entities.csv", index=False, encoding="utf-8")
    print(f"Extracted {len(df)} entities and saved to ../data/processed/entities.csv")
else:
    print("No entities found. Check if the text files contain readable abstracts.")
```
