```python
import sys
sys.path.append("..")
from kg.neo4j_loader import Neo4jLoader
import pandas as pd

loader = Neo4jLoader("bolt://localhost:7687", "neo4j", "password")  # change password if needed
loader.create_indexes()

entities = pd.read_csv("../data/processed/entities.csv")
unique = entities.drop_duplicates(subset=["text", "node_type"])
for _, row in unique.iterrows():
    attrs = {"id": row["text"], "name": row["text"], "type": row["node_type"], "ontology_id": row["ontology_id"]}
    loader.add_node(row["node_type"], attrs)

triples = pd.read_csv("../data/processed/triples.csv")
for _, row in triples.iterrows():
    attrs = {"relation_type": row["relation"], "evidence_score": 1.0, "sentence": row["sentence"], "source_file": row["source_file"]}
    loader.add_edge(row["head"], row["tail"], row["relation"], attrs)

print("Graph loaded successfully")
```
