```python
import pandas as pd
import os

# Load triples
triples_path = "../data/processed/triples.csv"
if not os.path.exists(triples_path):
    print(f"Error: {triples_path} not found.")
    print("Creating sample triples for demonstration...")
    sample = [
        ("Alzheimer's disease", "has_phenotype", "Memory loss"),
        ("APP", "causes", "Alzheimer's disease"),
        ("Donepezil", "treats", "Alzheimer's disease"),
        ("APOE", "associated_with", "Alzheimer's disease"),
    ]
    triples = pd.DataFrame(sample, columns=["head", "relation", "tail"])
else:
    triples = pd.read_csv(triples_path)
    if triples.empty:
        print("Triples file is empty. Using sample data.")
        triples = pd.DataFrame([
            ("Alzheimer's disease", "has_phenotype", "Memory loss"),
            ("APP", "causes", "Alzheimer's disease"),
        ], columns=["head", "relation", "tail"])

# Print as a simple text graph
print("\n" + "="*50)
print("DCE-KG Knowledge Graph (Text View)")
print("="*50)
for _, row in triples.iterrows():
    print(f"[{row['head']}] --{row['relation']}--> [{row['tail']}]")

# Also generate a simple DOT format (can be visualized with Graphviz online)
dot = "digraph DCEKG {\n"
for _, row in triples.iterrows():
    dot += f'    "{row["head"]}" -> "{row["tail"]}" [label="{row["relation"]}"];\n'
dot += "}"
print("\n" + "="*50)
print("Graphviz DOT code (copy to https://dreampuf.github.io/GraphvizOnline/ to visualize):")
print("="*50)
print(dot)

# Save DOT to file
with open("dcekg.dot", "w", encoding="utf-8") as f:
    f.write(dot)
print("\nDOT file saved as 'dcekg.dot'. Open it at https://dreampuf.github.io/GraphvizOnline/")
```
