```python
from pyvis.network import Network
import pandas as pd
import os

# Load triples
triples_path = "../data/processed/triples.csv"
if not os.path.exists(triples_path):
    print(f"Error: {triples_path} not found. Run 3_relation_extraction.ipynb first.")
    exit()

triples = pd.read_csv(triples_path)

if triples.empty:
    print("No triples found. Create sample data for visualization.")
    # Create dummy data for demonstration
    sample_data = [
        {"head": "Alzheimer's disease", "relation": "has_phenotype", "tail": "Memory loss"},
        {"head": "APP", "relation": "causes", "tail": "Alzheimer's disease"},
        {"head": "Donepezil", "relation": "treats", "tail": "Alzheimer's disease"},
    ]
    triples = pd.DataFrame(sample_data)
    print("Using sample data for visualization.")

# Create network
net = Network(height="750px", width="100%", bgcolor="#222222", font_color="white")

# Add nodes and edges
for _, row in triples.iterrows():
    # Add head node
    net.add_node(row["head"], label=row["head"], title=str(row.get("head_type", "Node")))
    # Add tail node
    net.add_node(row["tail"], label=row["tail"], title=str(row.get("tail_type", "Node")))
    # Add edge
    net.add_edge(row["head"], row["tail"], label=row["relation"], title=row.get("sentence", ""))

# Set physics options for better layout
net.set_options("""
var options = {
  "physics": {
    "enabled": true,
    "stabilization": {"iterations": 100}
  }
}
""")

# Save and open
output_file = "dcekg_viz.html"
try:
    net.show(output_file)
    print(f"Visualization saved as '{output_file}'. Open this file in your browser.")
    import webbrowser
    webbrowser.open(output_file)
except Exception as e:
    print(f"Error generating visualization: {e}")
    print("Try installing pyvis manually: pip install pyvis")
```
